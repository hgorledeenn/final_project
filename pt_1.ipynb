```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/registration_contacts.csv')
df.head()
```

```python
registration_ids = df['RegistrationID'].unique().tolist()
registration_ids[0:5]
```

```python
df_mdf = pd.read_csv('data/multiple_dwelling_registrations_all.csv')
df_mdf.head()
```

```python
df_mdf_filtered = df_mdf[df_mdf['RegistrationID'].isin(registration_ids)].reset_index(drop=True)
```

```python
df_mdf_filtered.head()
```

```python
df_merged = pd.merge(df_mdf_filtered, df, on='RegistrationID', how='left')
pd.set_option('display.max_columns', 30)
df_merged.head()
```

```python
df_merged.shape
```

```python
df_clean = df_merged[['RegistrationID', 'BuildingID', 'Boro', 'BoroID', 'HouseNumber', 'StreetName', 'Zip', 'Block', 'Lot', 'BIN', 'RegistrationContactID', 'CorporationName', 'BusinessHouseNumber', 'BusinessStreetName', 'BusinessApartment', 'BusinessCity', 'BusinessState', 'BusinessZip']]
df_clean.shape
```

```python
df_clean['house_num_string'] = df_clean['HouseNumber'].copy().astype('str')
df_clean['zip_int'] = df_clean['Zip'].copy().astype('int')
df_clean['zip_string'] = df_clean['zip_int'].copy().astype('str')
```

```python
df_clean.dtypes
```

```python
df_clean['building_address'] = df_clean['house_num_string'] + " " + df_clean['StreetName'] + ", NEW YORK, NY, " + df_clean['zip_string']
df_clean.head()
```

```python
df_clean['business_house_num_string'] = df_clean['BusinessHouseNumber'].copy().astype('str')
df_clean['business_zip_int'] = df_clean['BusinessZip'].copy().astype('int')
df_clean['business_zip_string'] = df_clean['business_zip_int'].copy().astype('str')
```

```python
has_apt = df_clean['BusinessApartment'].notna()

# 2. Build the full address using a conditional format
df_clean['business_address'] = (
    df_clean['business_house_num_string'] + ' '
    + df_clean['BusinessStreetName'] +
    np.where(has_apt, ' APARTMENT ' + df_clean['BusinessApartment'], '') + ', '
    + df_clean['BusinessCity'] + ', '
    + df_clean['BusinessState'] + ' '
    + df_clean['business_zip_string']
)

df_clean['business_address']
```

```python
df_clean.head()
```

```python
df_final = df_clean[['CorporationName', 'building_address', 'business_address', 'RegistrationID', 'BuildingID', 'Boro', 'BoroID', 'Block', 'Lot', 'BIN', 'RegistrationContactID']]
```

```python
df_final.to_csv('data/pt_1.csv', index=False)
```
